# 2. Исследование зависимостей и формулирование гипотез

```python
# Заменяем значения : 
# 'Yes', 'No', 'No phone service' , 'No internet service' на значения '1', '0', '2', '3'
# 'Male', 'Female' на значения '1', '2'
# 'DSL', 'Fiber optic', 'No' на значения '1', '2', '0'
# 'Month-to-month', 'One year', 'Two year' на значения '0','1', '2'
# 'Electronic check', 'Mailed check', 'Bank transfer (automatic)', 'Credit card (automatic) на значения '1', '2', '3', '4'

df = df.replace(['Yes', 'No', 'No phone service', 'No internet service'],[1, 0, 2, 3])
df = df.replace(['Male', 'Female'],[1, 2])
df = df.replace(['DSL', 'Fiber optic'],[1, 2])
df = df.replace(['Month-to-month', 'One year', 'Two year'],[0,1, 2])
df = df.replace(['Electronic check', 'Mailed check', 'Bank transfer (automatic)', 'Credit card (automatic)'],[1, 2, 3, 4])
```

```python
# Приводим столбцы к соответствующим форматам. 
df[['gender', 'Churn']] = df[['gender', 'Churn']].apply(pd.to_numeric)
```

Как мы видим линейная корреляция мала , поэтому выбираем преимущественно нелинейные модели. Для этого осуществим подбор лучших признаков 

```python
# Разбиваем имеющиеся данные на обучающие и тестовые

from sklearn.model_selection import (GridSearchCV,
                                     train_test_split,
                                     StratifiedKFold, cross_val_predict, cross_val_score)
import xgboost as xgb

X= df.drop(columns='Churn')
y = df['Churn']

X_train, X_test, y_train, y_test = train_test_split(X, 
                                                    y,
                                                    test_size=0.3,
                                                    stratify=y,
                                                    random_state=42)
```

```python
# Подготавливаем данные и обучаем модель для поиска часто испозуемых фичей
# Выводим график F-score фичей 

dtrain = xgb.DMatrix(X_train, y_train)
dtest = xgb.DMatrix(X_test, y_test)

xgb_params = {'objective': 'binary:logistic',
          'max_depth': 3,
          'eta': 0.1}

num_rounds = 60

xgb_model = xgb.train(xgb_params, dtrain, num_rounds)

xgb.plot_importance(xgb_model)
```

```python
# Получаем словарь с фичами и их значениями , сортируем и выподим в порядке возрастания

best_feature = xgb_model.get_score(importance_type='weight')
sort_best_feature = {k: best_feature[k] for k in sorted(best_feature, key=best_feature.get)}
sort_best_feature.keys()
```

```python
# Список лучших фичей

best_features = ['InternetService', 'MultipleLines', 'OnlineBackup', 'PaperlessBilling', 'OnlineSecurity', 'PaymentMethod', 'Contract', 'tenure', 'TotalCharges', 'MonthlyCharges']
```

# 3. Построение моделей для прогнозирования оттока

```python
from sklearn.ensemble import GradientBoostingClassifier, RandomForestClassifier
from sklearn.svm import SVC
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import roc_auc_score
from sklearn.model_selection import cross_val_predict
import xgboost as xgb
from sklearn.metrics import  precision_score, recall_score
from xgboost import XGBClassifier
```

```python
# наиболее важные признаки

X_train = X_train[best_features]
X_test = X_test[best_features]
```

```python
# параметры базовых алгоритмов

gbc_params = {'learning_rate': np.arange(0.1, 0.6, 0.1)} # GradientBoostingClassifier

rfc_params = {'n_estimators': range(10, 100, 10), # RandomForestClassifier
              'min_samples_leaf': range(1, 5)}

svc_params = {'C': np.arange(0.1, 1, 0.2)} # SVC

lr_params = {'C': np.arange(0.5, 1, 0.1)} # LogisticRegression

skf = StratifiedKFold(n_splits=5, shuffle=True, random_state=42)
```

```python
# создание словаря с параметрами моделей для их оптимизации 

best_params = {GradientBoostingClassifier(random_state=42): gbc_params,
               RandomForestClassifier(random_state=42): rfc_params, 
               SVC(random_state=42): svc_params,
               LogisticRegression(random_state=42): lr_params
               }
```

```python
# создаём список моделей с оптимизированными параметрами 

models_best_params = []

for model, param in best_params.items():
  grid = GridSearchCV(model, param, scoring='roc_auc', cv=skf, n_jobs=-1)
  grid.fit(X_train, y_train)
  models_best_params.append(model.set_params(**grid.best_params_))

# обучаем модели и формируем из н

model_gbc = models_best_params[0].fit(X_train, y_train)
model_rfc = models_best_params[1].fit(X_train, y_train)
model_svc = models_best_params[2].fit(X_train, y_train)
model_lr = models_best_params[3].fit(X_train, y_train)

models = [model_gbc, model_rfc, model_svc, model_lr]
```

```python
models = [model_gbc, model_rfc, model_svc, model_lr]
```

```python
# Проводим оценку качеста по roc-auc и выводим результаты

model_name = ['GradientBoostingClassifier', 'RandomForestClassifier', 'SVC', 'LogisticRegression']
score = {}

for n, model in enumerate(models):
    roc_auc = roc_auc_score(y_test, model.predict(X_test))
    score[model_name[n]] = roc_auc
    print(f'{model_name[n]} : {roc_auc}')
```

```python
# Попробуем использовать алгоритв XGBoost

dtrain = xgb.DMatrix(X_train, y_train)
dtest = xgb.DMatrix(X_test, y_test)

xgb_params = {'objective': 'binary:logistic',
          'max_depth': 3,
          'eta': 0.1}

num_rounds = 60
xgb_model = xgb.train(xgb_params, dtrain, num_rounds)

roc_score = roc_auc_score(y_test, xgb_model.predict(dtest))
score['XGBoost'] = roc_score
print(f'XGBoost : {roc_score}')
```

```python
# Построим модель с обновлёнными параметрами , также зададим порог вероятностного выхода модели

xgb_model = xgb.train(xgb_params, dtrain, num_rounds)
roc_score = roc_auc_score(y_test > 0.45, xgb_model.predict(dtest)> 0.45)
print(f'XGBoost с зададим порог вероятностного выхода модели в 45%: {roc_score}')
roc_score = roc_auc_score(y_test > 0.2, xgb_model.predict(dtest)> 0.2)
print(f'XGBoost с зададим порог вероятностного выхода модели в 20%: {roc_score}')
```

```python
# Построим мета-модель , в качестве мета-алгоритма будем использовать XGBClassifier
# Найдём для него лучшие параметры,

xgbс_params = {'n_estimators': range(10, 100, 5),
              'eta': np.arange(0.1, 1., .1),
              'min_child_weight': range(1, 10, 1),
              'subsample': np.arange(0.1, 1., 0.2)}

meta_mtrx = np.empty((y_train.shape[0], len(models)))

for i, model in enumerate(models):
    meta_mtrx[:, i] = model.predict(X_train)
    
xgbc = XGBClassifier(random_state=17)
grid_xgb = GridSearchCV(xgbc, param_grid=xgb_params, scoring='roc_auc', cv=skf)
model_xgbс = grid_xgbс.fit(meta_mtrx, y_train)
model_xgbс.best_params_
```

```python
# Мета-алгоритм с новыми параметрами

model_xgbс = XGBClassifier(random_state=42, eta=0.8, min_child_weight=4, n_estimators=65, subsample=0.30000000000000004)
```

```python
# поиск оптимальных параметров мета-модели
meta_alg_params = {'max_depth': [2], 'n_estimators': [80], 'n_jobs' : [1], 'random_state': [42]}

meta_model = GridSearchCV(model_xgbс, meta_alg_params, scoring='roc_auc', cv=5)
# обучение мета_модели
meta_model.fit(meta_mtrx, y_train)
```

```python
# Проводим оценку качеста мета-модели по roc_auc

meta_mtrx_test = np.empty((y_test.shape[0], len(models)))         
for i, model in enumerate(models):
    meta_mtrx_test[:, i] = model.predict(X_test)
            
roc_score = roc_auc_score(y_test, meta_model.predict(meta_mtrx_test))
score['meta_model'] = roc_score
roc_score
```
